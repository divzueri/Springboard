## Summary

I. Raw Data:<br>
a) Steam dataset: Contains information on about 27k games on the steam store released before 2019.<br>
b) Tags dataset: 29k steam games and how many time each tag (~370 total) was used to describe it.<br>
<br>
II. Checked for missing values and ensured columns were correct data type<br>
<br>
III. Numerical Columns<br>
    - Created aggregate values, such as averaging the interval value for Owners, multiplied this by price to predict revenue, etc.<br>
    - For release date, created separate columns for season of release and years since release<br>
<br>
IV. Categorical Columns<br>
    - Separate and reformat tag-based columns so that each tag of interest is a unique column<br>
    - For genre-based tags, identify the top 20 most commonly used tags.<br>
    - For game-type tags, group similar tags (e.g. multiplayer, online multiplayer, local multiplayer = one tag)<br>
<br>
V. Merge dataframes, keep columns of interest in final cleaned dataframe<br>
<br>
Target Vars:<br>
    1) Revenue (average owners * cost)<br>
    2) % Positive Ratings (Pos / (Pos + Neg))

## 1.3 Ensure Variables Are Correct Type

```python
# Convert release_date to datetime
steam_df['release_date'] = [datetime.strptime(date_str, '%Y-%m-%d') for date_str in steam_df['release_date']]
# Convert appid to string
steam_df['appid'] = steam_df['appid'].astype(str)
# Convert english to bool
steam_df['english'] = steam_df['english'].astype(bool)
```

## 1.4 Explore Numeric Variables

```python
# Create new column averaging interval of owners
steam_df['avg_owners'] = [((int(a)+int(b))/2) for a,b, in [i for i in [x for x in steam_df['owners'].str.split('-')]]]
steam_df[['owners', 'avg_owners']].head()
```

```python
# Create revenue column
steam_df['revenue'] = steam_df['avg_owners']*steam_df['price']
steam_df[['avg_owners', 'price', 'revenue']].head()
```

```python
# Create columns calculating total ratings
steam_df['total_ratings'] = steam_df['positive_ratings']+steam_df['negative_ratings']

# Create column calculating percentage of positive ratings
steam_df['perc_pos_ratings'] = steam_df['positive_ratings']/steam_df['total_ratings']
steam_df[['positive_ratings', 'negative_ratings', 'total_ratings', 'perc_pos_ratings']].head()
```

### 1.6.1 Clean 'Tag-Based' Columns

```python
# Lower-case 'tag-based' categorical features
steam_df['categories'] = steam_df['categories'].str.lower()
steam_df['genres'] = steam_df['genres'].str.lower()
steam_df['steamspy_tags'] = steam_df['steamspy_tags'].str.lower()
```

```python
# Convert 'tag-based' categorical features from strings to lists in new columns
steam_df['categories_list'] = steam_df['categories'].str.split(';')
steam_df['genres_list'] = steam_df['genres'].str.split(';')
steam_df['steamspy_list'] = steam_df['steamspy_tags'].str.split(';')
```

```python
# Sort each list in alphabetical order
steam_df['categories_list'] = [sorted(i) for i in steam_df['categories_list']]
steam_df['genres_list'] = [sorted(i) for i in steam_df['genres_list']]
steam_df['steamspy_list'] = [sorted(i) for i in steam_df['steamspy_list']]
```

### a) Categories

```python
# Create list of all values from categories column and their counts
categories = sorted(steam_df['categories_list'].explode())
cat_counts = Counter(categories)
```

```python
cat_counts.most_common()
```

Several of these can be combined. Others might not be that useful for analysis

 - Multiplayer = multi-player, online multi-player, local multi-player, cross-platform multiplayer
 - Co-op = co-op, online co-op, local co-op
 - Full controller support/partial controller support
 - In app purchases
 - Steam achievements
 - Steam trading cards
 - Steam cloud

```python
# Create categories df
cat_df = steam_df[['appid','categories_list']]
cat_df.head()
```

```python
#Lists of categories to group together
multiplayer = ['multi-player','online multi-player','local multi-player','cross-platform multiplayer', 'co-op', 'online co-op', 'local co-op', 'mmo']
coop = ['co-op', 'online co-op', 'local co-op']
cont_supp = ['full controller support', 'partial controller support']
steam_achiev = ['steam achievements']
steam_tc = ['steam trading cards']
steam_cloud = ['steam cloud']
```

```python
# List of cat lists + col names
col_names = ['multi', 'co-op', 'cont_supp', 'steam_achiev', 'steam_tc', 'steam_cloud']
cat_lists = [multiplayer, coop, cont_supp, steam_achiev, steam_tc, steam_cloud]
```

```python
def create_new_col(new_col, lst):
    """Checks if each game matches with any item in input list,
    then creates new column with according value"""
    cat_df[new_col] = ""
    for idx, game in enumerate(cat_df['categories_list']):
        if any(i in game for i in lst) is True:
            cat_df[new_col][idx] = 1
        else:
            cat_df[new_col][idx] = 0
```

```python
# Apply function to each category list
[create_new_col(x,y) for x,y in zip(col_names, cat_lists)]
```

```python
cat_df.head(20)
```

```python
[cat_df[i].value_counts() for i in cat_df.columns[2:]]
```

 - Less than a quarter of games are multi-player
 - Under half of all multi-player games are co-op
 - A third of games have some fort of controller support
 - A little over haf of all games offer steam achievements
 - Under a third of all games offer steam trading cards
 - Under a third have steam cloud

```python
# Remove categories_list column
catclean_df = cat_df.drop('categories_list', axis=1)
catclean_df.head()
```

# 2. Steam Tags Dataset

## 2.1 Load Data

```python
tags_df = pd.read_csv('../raw_data/steamspy_tag_data.csv')
```

```python
tags_df.shape
```

```python
tags_df.head()
```

## 2.2 Check for Missing Values

```python
tags_df.isna().sum().sort_values(ascending=False)
```

## 2.3 Rewrite df to Binary Values

```python
# Copy original df
tags_bool = tags_df
```

```python
# Convert appid to string
tags_bool['appid'] = tags_bool['appid'].astype(str)
```

```python
# Make a list of all columns excluding ID column
col_names = tags_bool.iloc[:,1:].columns
```

```python
col_names
```

```python
# Set all columns to binary values (1 = at least one tag exists for this genre)
tags_bool[col_names] = np.where(tags_bool[col_names]==0, 0, 1)
```

```python
tags_bool.head()
```

## 2.4 Tags Per Game

```python
# Add column listing how many tags each game has
tags_bool['tags_per_game'] = tags_bool[col_names].sum(axis=1)
```

```python
tags_per_game = tags_bool[['appid','tags_per_game']]
tags_per_game.head()
```

```python
tags_per_game[['tags_per_game']].describe()
```

Average tags/game is 8-9. Max is 22.

## 2.5 Top Tags

```python
# Transpose df
tags_T = tags_bool.T
tags_T.head()
```

```python
# Remove aggregate 'tags_per_game' row
tags_T.drop('tags_per_game', axis=0, inplace=True)
```

```python
# Reset index
tags_T.reset_index(inplace=True)
tags_T.head()
```

```python
# Reassign first row as column names
tags_T.columns = tags_T.iloc[0]
tags_T.drop([0],inplace=True)
```

```python
# Rename first column
tags_T.rename(columns={'appid':'tag_name'}, inplace=True)
```

```python
tags_T.head()
```

```python
# Create sum column (games per tag)
tags_T['all_games'] = tags_T[tags_T.iloc[:,1:].columns].sum(axis=1)
```

```python
tags_T[['tag_name', 'all_games']].head()
```

```python
# Sort games by number of tags
tags_sorted = tags_T.sort_values('all_games', ascending=False)
tags_sorted[['tag_name','all_games']].head()
```

```python
# Reset index
tags_sorted.reset_index(drop=True, inplace=True)
```

```python
# Add rank column
tags_sorted['rank'] = tags_sorted.index + 1
```

```python
tags_sorted[['tag_name','all_games','rank']].head(40)
```

Some of these tags represent game category more so than game genre

```python
# View summary stats for all tags
tags_sorted['all_games'].describe()
```

```python
# Distribution of games per tag
plt.hist(tags_sorted['all_games']);
plt.title('Distribution of games per tag');
plt.xlabel('Number of games using tag');
plt.ylabel('Count');
```

Most of these tags are used by very few games

```python
# Distribution of the top 30 tags
plt.bar(tags_sorted['tag_name'][:30],tags_sorted['all_games'][:30]);
plt.title('Distribution of games per tag for the top 30 tags');
plt.xlabel('Top 20 Tags');
plt.ylabel('Number of games using tag')
plt.xticks(rotation=90);
```

```python
tags_sorted['all_games'][:30].describe()
```

```python
# Create top 20 tags df to merge back with steam_df
# Will later drop tags that are already represented in cat_df
top_20 = tags_sorted[:23]
```

```python
top_20.drop(columns=['all_games','rank'],inplace=True)
```

```python
top_20.head()
```

```python
top_20 = top_20.T
```

```python
top_20.head()
```

```python
top_20.rename_axis("appid",inplace=True)
```

```python
# Reset index
top_20.reset_index(inplace=True)
# Reassign first row as column names
top_20.columns = top_20.iloc[0]
top_20.drop([0],inplace=True)
top_20.rename(columns={'tag_name':'appid'}, inplace=True)
```

```python
top_20.columns
```

```python
# Drop singleplayer, multiplayer, and early_access
top_20.drop(columns=['singleplayer', 'early_access', 'multiplayer'], inplace=True)
```

# 3. Merge Dataframes

Here I'll be merging catclean_df, tags_per_game, and top_20 to steam_df

```python
# Check shape of new dfs
catclean_df.shape, top_20.shape, tags_per_game.shape
```

```python
# Merge tags_per_game
steam_df = steam_df.merge(tags_per_game, on='appid', how='left')
```

```python
# Merge catclean_df
steam_df = steam_df.merge(catclean_df, on='appid', how='left')
```

```python
# Merge top_20 df
steam_df = steam_df.merge(top_20, on='appid', how='left')
```
