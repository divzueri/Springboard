```python
import json
import pandas as pd
import numpy

import seaborn as sns
import matplotlib.pyplot as plt

from datetime import datetime
```

# I. Open and Set Up Data

```python
f = open('logins.json')
```

```python
data = json.load(f)
```

```python
data.keys()
```

```python
data['login_time'][:5]
```

```python
# Create df and convert login_time to datetime type
df = pd.DataFrame(data)
df['login_time'] = pd.to_datetime(df['login_time'])

# Move datetime to index and add numeric col for aggregation
df.set_index('login_time', inplace=True)
df['count'] = 1
df.sort_values('login_time').head()
```

```python
# Aggregate login counts into 15 min intervals

df_interval = df.resample('15min').sum()
df_interval.head()
```

```python
# Add datetime cols

df_interval['hour'] = df_interval.index.hour
df_interval['weekday'] = df_interval.index.day_name()
df_interval['month'] = df_interval.index.month_name()

df_interval.head()
```

```python
df_interval.isna().sum()
```

```python
df_interval.describe()
```

```python
month_idx = ['January', 'February', 'March', 'April']

df_hourmonth = df_interval.groupby(['hour', 'month']).sum().reindex(month_idx, level='month').reset_index()

fig, ax = plt.subplots(figsize=(15,5))

sns.lineplot(data=df_hourmonth, x='hour', y='count', hue='month', ax=ax);
ax.set_title('Total User logins by hour and month');
ax.legend(loc='upper right', bbox_to_anchor=(1.15, 1.15));
```

Overall logins decrease from Jan to April, but the hourly patterns remain similar. This might be because users might be more willing to walk in warmer weather.

```python
df_daymonth = df_interval.groupby(['weekday', 'month'])[['count']].sum().reindex(
    month_idx, level='month').reindex(day_index, level='weekday').reset_index()

fig, ax = plt.subplots(figsize=(15,5))

sns.lineplot(data=df_daymonth, x='weekday', y='count', hue='month', ax=ax);
ax.set_title('Total User logins by day of week and month');
ax.legend(loc='upper right', bbox_to_anchor=(1.15, 1.15));
```

Similar to the previous graph, logins steadily decrease from Jan to April. The weekly pattern is somewhat consistent across months, although only Jan sees a spike in logins on Sundays. Tuesdays also seem to experience more fluctuation by month.
